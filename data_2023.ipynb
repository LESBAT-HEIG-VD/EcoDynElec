```python
import pandas as pd
from ecodynelec.preprocessing.enr_residual_utils import load_pronovo_file,load_all_pronovo_files,reorganize_enr_data,read_enr_data_from_energy_charts,read_enr_data_from_pronovo
```

```python
#Test these two functions carefully to understand how they're working 
#They are the root of the whole loading step 
df=load_all_pronovo_files('D:/Model/ecd_enr_model/',['pronovo_data/prod_2023'], types=['Solar'], verbose=True)
#df=load_pronovo_file(r"D:\Model\ecd_enr_model\prod_2023\202303_CH_Total_Quartal_def.csv",['Wind'],verbose=True)
df.index=pd.to_datetime(df.index,dayfirst=True)
```

```python
#Load pronovo_data 
#Read the function description 
pronovo_data=read_enr_data_from_pronovo('D:\Model\ecd_enr_model',verbose=True)
```

```python
#Load energy_charts data
#Read the function description 
ec_data=read_enr_data_from_energy_charts('D:\Model')
```

```python
#And then load the final dataframe which will be exported to ecodynelec 
final_enr_data=reorganize_enr_data(pronovo_data, ec_data, verbose = False)
```

```python
old_enr_data = pd.read_csv('D:\Model\ecd_enr_model\export\enr_prod_2016-2022_completed.csv', index_col=0,parse_dates=[0])
```

```python
#Update the data to export it to ecodynelec and then save it 
from datetime import timedelta
export_to_ecodyn=pd.concat([old_enr_data.loc[:(final_enr_data.index[0] - timedelta(hours=1))], final_enr_data], axis=0)
#Be sure that there is no duplicates in your dataframe after concatanation 
duplicate_indices = export_to_ecodyn.index[export_to_ecodyn.index.duplicated()]
duplicate_indices
```
